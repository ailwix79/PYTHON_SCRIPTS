### PRÁCTICA MULTITHREADING 1.1 (CÓDIGO SERVIDOR)

```python
import threading
import socket
import pickle

def input_eval_loop():
    while True:
        try:
            port = int(input("Port number> "))

            if (port <= 0 or port >= 65535):
                print("Port must be a number from 0 to 65535")
            else:
                break
        except ValueError:
            print("Port must be a number from 0 to 65535")
            continue
            
    return port
def prepare_socket(port):
    try:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.bind(("127.0.0.1",port))
        sock.listen(1)
        return sock
    except:
        print("Socket creation failed")

def create_client_thread(conn, client_address):
    try:
        data = pickle.loads(conn.recv(1024))
        
        print(str(data[0]) + ": Received message from client " + str(client_address) + "> " + str(data[1]))
        
        conn.sendall(pickle.dumps(data))
        conn.close()
        
        print(str(data[0]) + ": Closing socket")
        print("Waiting for client...")
    except:
        print("Connection failed")
    
def server_main(sock):         
    print("Waiting for client...")
    while True:
        try:
            conn, client_address = sock.accept() 
            threading.Thread(target=create_client_thread, args=(conn, client_address)).start()
        except:
            print("Client attempted to join but something has failed. Falling back to waiting mode")
            continue

if __name__ == "__main__":
    port = input_eval_loop()
    sock = prepare_socket(port)
    server_main(sock)
    
```

### PRÁCTICA MULTITHREADING 1.2 (CÓDIGO SERVIDOR)

```python
import threading
import socket
import pickle

global lock
lock = threading.Lock()

def input_eval_loop():
    while True:
        try:
            port = int(input("Port number> "))

            if (port <= 0 or port >= 65535):
                print("Port must be a number from 0 to 65535")
            else:
                break
        except ValueError:
            print("Port must be a number from 0 to 65535")
            continue
            
    return port
def prepare_socket(port):
    try:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.bind(("127.0.0.1",port))
        sock.listen(1)
        return sock
    except:
        print("Socket creation failed")

def create_client_thread(conn, client_address):
    try:
        data = pickle.loads(conn.recv(1024))
        
        with lock:
            print(str(data[0]) + ": Received message from client " + str(client_address) + "> " + str(data[1]))
        
        conn.sendall(pickle.dumps(data))
        conn.close()
        
        with lock:
            print(str(data[0]) + ": Closing socket")
            print("Waiting for client...")
    except:
        with lock:
            print("Connection failed")
    
def server_main(sock):         
    print("Waiting for client...")
    while True:
        try:
            conn, client_address = sock.accept() 
            threading.Thread(target=create_client_thread, args=(conn, client_address)).start()
        except:
            print("Client attempted to join but something has failed. Falling back to waiting mode")
            continue

if __name__ == "__main__":
    port = input_eval_loop()
    sock = prepare_socket(port)
    server_main(sock)
```
